```python


import matplotlib.pyplot as plt
from matplotlib import rc

# Open and read the file
file_path = '/home/nicolas/Computacion-Paralela-2025-FaMAF/Lab1/Experimentos/exp0.txt'
with open(file_path, 'r') as file:
    lines = file.readlines()

x_values = []
y_values = []

for line in lines:
    line = line.strip()
    if not line or line.startswith('Flags:') or line.startswith('#'):
        continue
    
    parts = line.split()
    if len(parts) >= 2:
        try:
            x = float(parts[0])
            y = float(parts[1])
            x_values.append(x)
            y_values.append(y)
        except ValueError:
            continue



# Data processing code remains the same...

if not x_values:
    print("No valid data found in the file!")
else:
    # Set up LaTeX rendering with fallback
    try:
        rc('text', usetex=True)
        rc('font', family='serif')
        latex_available = True
    except:
        latex_available = False
        print("LaTeX not available - using standard text rendering")

    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, 'b-', linewidth=2)
    
    if latex_available:
        plt.xlabel(r'$\nu$', fontsize=12)
        plt.ylabel(r'$K\nu/s$', fontsize=12)
        info_text = (
            r"\begin{tabular}{ll}"
            r"Compilador: & gcc \\"
            r"Flags: & \texttt{-g -std=c11 -ffast-math} \\"
            r" & \texttt{-Wall -Wextra -Werror -O3} \\"
            r" & \texttt{-march=native} \\"
            r"Ejecución: & \texttt{taskset -c 0,1 ./headless} \\"
            r"Compilación: & Makefile"
            r"\end{tabular}"
        )
    else:
        plt.xlabel('ν', fontsize=12)
        plt.ylabel('Kν/s', fontsize=12)
        info_text = (
            "Compilador: gcc\n"
            "Flags: -g -std=c11 -ffast-math\n"
            "       -Wall -Wextra -Werror -O3\n"
            "       -march=native\n"
            "Ejecución: taskset -c 0,1 ./headless\n"
            "Compilación: Makefile"
        )

    plt.title('Performance Results in Atom', fontsize=14)
    plt.grid(True, alpha=0.3)

    plt.text(0.85, 0.85, info_text,
             transform=plt.gca().transAxes,
             fontsize=10,
             verticalalignment='top',
             horizontalalignment='right',
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray', boxstyle='round'))

    plt.tight_layout()
    plt.show()
```
