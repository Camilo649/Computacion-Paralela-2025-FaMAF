```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file (assuming space-separated)
# If your file has a different delimiter, adjust `sep` (e.g., sep='\t' for tabs)
df = pd.read_csv('/home/nicolas/Computacion-Paralela-2025-FaMAF/Lab1/Benchmarks/streamResults.txtaa', sep='\s+', skiprows=1, header=None, 
                 names=['Function', 'Best Rate MB/s', 'Avg time', 'Min time', 'Max time'])

# Remove ':' from Function names (e.g., "Copy:"  "Copy")
df['Function'] = df['Function'].str.replace(':', '')

# --- Plot 1: Best Rate MB/s (Bar Plot) ---
plt.figure(figsize=(8, 5))
bars = plt.bar(df['Function'], df['Best Rate MB/s'], color='skyblue')
plt.title('Best Rate (MB/s) by Function', fontsize=14)
plt.xlabel('Function', fontsize=12)
plt.ylabel('Best Rate (MB/s)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, 
             f'{height:,.0f}', ha='center', va='bottom')

plt.show()

# --- Plot 2: Time Comparison (Grouped Bar Plot) ---
plt.figure(figsize=(10, 6))
width = 0.25
x = range(len(df['Function']))

# Plot Avg, Min, Max times side by side
plt.bar(x, df['Avg time'], width, label='Avg time', color='lightgreen')
plt.bar([i + width for i in x], df['Min time'], width, label='Min time', color='orange')
plt.bar([i + 2*width for i in x], df['Max time'], width, label='Max time', color='pink')

plt.xticks([i + width for i in x], df['Function'])
plt.title('Time Comparison by Function', fontsize=14)
plt.xlabel('Function', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
threads = np.array([4, 16, 32])
mphotons_per_sec = np.array([7.832625, 30.333589, 59.541240])
colors = ['#00e6b8', '#f5deb3', 'black']

# Posiciones equidistantes para que las barras estn juntas
x_positions = np.arange(len(threads))

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(x_positions, mphotons_per_sec, color=colors, edgecolor='gray')

# Etiquetas
plt.xticks(x_positions, threads)
plt.xlabel("Cantidad de hilos")
plt.ylabel("Millones de fotones por segundo")
plt.title("Rendimiento vs Cantidad de Hilos")

# Mostrar valores encima de las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.5,
             f"{height:.4f}", ha='center', va='bottom')

# Agregar porcentaje de mejora entre barras
for i in range(1, len(mphotons_per_sec)):
    prev = mphotons_per_sec[i - 1]
    curr = mphotons_per_sec[i]
    pct = (curr - prev) / prev * 100
    mid_x = (x_positions[i - 1] + x_positions[i]) / 2
    mid_y = (prev + curr) / 2

    # Lnea o flecha indicativa (opcional: reemplaz por annotate con arrowprops si quers una flecha real)
    plt.annotate(f"+{pct:.1f}%", 
                 xy=(mid_x, mid_y), 
                 xytext=(mid_x, mid_y + 5),
                 ha='center', 
                 arrowprops=dict(arrowstyle='->', color='gray'))

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.figtext(0.5, -0.05, "Simulacin con 16G fotones", ha="center", fontsize=10)
plt.show()
```
