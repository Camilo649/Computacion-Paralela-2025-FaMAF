```python
# Importing all necesary libraries for plotting
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the CSV file
df = pd.read_csv('/home/augusto/Computacion-Paralela-2025-FaMAF/tiny_mc/results.csv')

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['PHOTONS'], df['Kphotons/seg'], marker='o', linestyle='-', color='b')
plt.xlabel('PHOTONS')
plt.ylabel('Kphotons/seg')
plt.title('Performance vs PHOTONS')
plt.grid(True)
plt.savefig('plot.png')  # Save the plot as an image
plt.show()
```
