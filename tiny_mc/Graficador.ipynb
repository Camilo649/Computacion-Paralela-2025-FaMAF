```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file (assuming space-separated)
# If your file has a different delimiter, adjust `sep` (e.g., sep='\t' for tabs)
df = pd.read_csv('/home/nicolas/Computacion-Paralela-2025-FaMAF/Lab1/Benchmarks/streamResults.txtaa', sep='\s+', skiprows=1, header=None, 
                 names=['Function', 'Best Rate MB/s', 'Avg time', 'Min time', 'Max time'])

# Remove ':' from Function names (e.g., "Copy:" → "Copy")
df['Function'] = df['Function'].str.replace(':', '')

# --- Plot 1: Best Rate MB/s (Bar Plot) ---
plt.figure(figsize=(8, 5))
bars = plt.bar(df['Function'], df['Best Rate MB/s'], color='skyblue')
plt.title('Best Rate (MB/s) by Function', fontsize=14)
plt.xlabel('Function', fontsize=12)
plt.ylabel('Best Rate (MB/s)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, 
             f'{height:,.0f}', ha='center', va='bottom')

plt.show()

# --- Plot 2: Time Comparison (Grouped Bar Plot) ---
plt.figure(figsize=(10, 6))
width = 0.25
x = range(len(df['Function']))

# Plot Avg, Min, Max times side by side
plt.bar(x, df['Avg time'], width, label='Avg time', color='lightgreen')
plt.bar([i + width for i in x], df['Min time'], width, label='Min time', color='orange')
plt.bar([i + 2*width for i in x], df['Max time'], width, label='Max time', color='pink')

plt.xticks([i + width for i in x], df['Function'])
plt.title('Time Comparison by Function', fontsize=14)
plt.xlabel('Function', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
photons = np.array([134217728, 1073741824, 17179869184])
mphotons_per_sec = np.array([82.300421, 83.560759, 84.741091])
colors = ['#00e6b8', '#f5deb3', 'black']

# Posiciones personalizadas (e.g., 0, 1, 2 para que estén equidistantes)
x_positions = np.arange(len(photons))

# Etiquetas como potencias de 2
xtick_labels = [f"$2^{{{int(np.log2(p))}}}$" for p in photons]

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(x_positions, mphotons_per_sec, color=colors, edgecolor='gray')

# Ejes y etiquetas
plt.xticks(x_positions, xtick_labels)
plt.xlabel("Cantidad de fotones")
plt.ylabel("Millones de fotones por segundo")
plt.title("Rendimiento vs Cantidad de Fotones")

# Mostrar valores encima de las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.3,
             f"{height:.4f}", ha='center', va='bottom')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.figtext(0.5, -0.05, "Simulación con 48 hilos en Atom", ha="center", fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
threads = np.array([4, 16, 32])
mphotons_per_sec = np.array([7.832625, 30.333589, 59.541240])
colors = ['#00e6b8', '#f5deb3', 'black']

# Posiciones equidistantes para que las barras estén juntas
x_positions = np.arange(len(threads))

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(x_positions, mphotons_per_sec, color=colors, edgecolor='gray')

# Etiquetas
plt.xticks(x_positions, threads)
plt.xlabel("Cantidad de hilos")
plt.ylabel("Millones de fotones por segundo")
plt.title("Rendimiento vs Cantidad de Hilos")

# Mostrar valores encima de las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.5,
             f"{height:.4f}", ha='center', va='bottom')

# Agregar porcentaje de mejora entre barras
for i in range(1, len(mphotons_per_sec)):
    prev = mphotons_per_sec[i - 1]
    curr = mphotons_per_sec[i]
    pct = (curr - prev) / prev * 100
    mid_x = (x_positions[i - 1] + x_positions[i]) / 2
    mid_y = (prev + curr) / 2

    # Línea o flecha indicativa (opcional: reemplazá por annotate con arrowprops si querés una flecha real)
    plt.annotate(f"+{pct:.1f}%", 
                 xy=(mid_x, mid_y), 
                 xytext=(mid_x, mid_y + 5),
                 ha='center', 
                 arrowprops=dict(arrowstyle='->', color='gray'))

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.figtext(0.5, -0.05, "Simulación con 16G fotones", ha="center", fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Datos
chunk_sizes = np.array([
    1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096,
    8192, 16384, 32768, 65536, 131072, 262144, 524288, 1048576,
    2097152, 4194304, 8388608, 16777216, 33554432, 67108864,
    134217728, 268435456
])

dynamic1 = np.array([
    83.953346, 84.293460, 84.510214, 84.601725, 84.630967, 84.684381, 84.677245,
    84.720270, 84.713470, 84.710511, 84.700539, 84.700279, 84.684183, 84.696816,
    84.672655, 84.708275, 84.673648, 84.647728, 84.488375, 84.222973, 84.133254,
    82.538417, 82.539572, 76.268204, 76.615082, 58.614446, 58.539966, 29.929712,
    15.743853
])

dynamic2 = np.array([
    82.456003, 84.726253, 84.731052, 84.732093, 84.741091, 84.541128, 84.735725,
    84.542932, 84.732791, 84.739279, 84.714219, 84.731971, 84.549255, 83.735400,
    84.651685, 84.731121, 84.651844, 84.579777, 84.646237, 84.249853, 83.288105,
    83.629999, 80.515491, 81.148060, 75.094619, 58.619519, 58.550895, 29.918366,
    15.666589
])

static_chunk = 4573943
static_value = 84.550646071

# Mejores índices
best_dyn1_idx = np.argmax(dynamic1)
best_dyn2_idx = np.argmax(dynamic2)

# Gráfico
bar_height = 0.25
indices = np.arange(len(chunk_sizes))
plt.figure(figsize=(12, 9))

# Dinámica
for i, val in enumerate(dynamic1):
    edgecolor = 'red' if i == best_dyn1_idx else 'gray'
    plt.barh(indices[i] - bar_height, val, height=bar_height, color='#f5deb3', edgecolor=edgecolor)
    if i == best_dyn1_idx:
        plt.text(val + 0.1, indices[i] - bar_height, f"{val:.2f}", va='center', fontsize=9)

# Guiada
for i, val in enumerate(dynamic2):
    edgecolor = 'blue' if i == best_dyn2_idx else 'gray'
    plt.barh(indices[i], val, height=bar_height, color='#00e6b8', edgecolor=edgecolor)
    if i == best_dyn2_idx:
        plt.text(val + 0.1, indices[i], f"{val:.2f}", va='center', fontsize=9)

# Estática
static_pos = np.searchsorted(chunk_sizes, static_chunk)
plt.barh(static_pos + bar_height, static_value, height=bar_height, color='black')
plt.text(static_value + 0.1, static_pos + bar_height, f"{static_value:.2f}", va='center', fontsize=9)

# Leyenda sin usar barras fuera del gráfico
legend_patches = [
    mpatches.Patch(color='#f5deb3', label='Dinámica'),
    mpatches.Patch(color='#00e6b8', label='Guiada'),
    mpatches.Patch(color='black', label='Estática'),
]
plt.legend(handles=legend_patches)

# Etiquetas y formato
plt.xlabel('Millones de fotones por segundo')
plt.ylabel('CHUNK SIZE')
plt.title('Rendimiento según CHUNK SIZE y tipo de planificación')
plt.yticks(indices, [f"{c:,}" for c in chunk_sizes])
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.figtext(0.5, -0.04, "Simulación con 16G fotones", ha="center", fontsize=10)

plt.show()
```
